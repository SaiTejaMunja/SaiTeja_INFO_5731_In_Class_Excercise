Question 3 (10 points): Use any of the feature selection methods mentioned in this paper "Deng, X., Li, Y., Weng, J., & Zhang, J. (2019). Feature selection for text classification: A review. Multimedia Tools & Applications, 78(3)."

Select the most important features you extracted above, rank the features based on their importance in the descending order.

```python
# Feature selection methods mentioned in the paper are Mutual Information, Chi-Square, Information Gain, Document Frequency, and Term Frequency-Inverse Document Frequency (TF-IDF).

# You code here (Please add comments in the code):

#importing necessary libraries and packages

import numpy as np
from sklearn.feature_selection import mutual_info_classif, chi2, f_classif
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Sample data
corpus = ["The product is really good.",
          "I love it!",
          "Not bad, but could be better.",
          "This is the worst product ever.",
          "I'm neutral about it."]

# Sample labels
labels = ["positive", "positive", "negative", "negative","neutral"]

# Vectorizing the text data using TF-IDF and CountVectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf_features = tfidf_vectorizer.fit_transform(corpus)

count_vectorizer = CountVectorizer()
count_features = count_vectorizer.fit_transform(corpus)

# Converting labels to numerical values (positive: 1, neutral: 0, negative: -1)
label_mapping = {"positive": 1, "neutral": 0, "negative": -1}
numerical_labels = [label_mapping[label] for label in labels]

# Creating Function to rank features using various methods
def rank_features(feature_matrix, labels):
    # Mutual Information
    #mi_scores = mutual_info_classif(feature_matrix, labels, discrete_features='auto')

    # Chi-Square
    chi2_scores, _ = chi2(feature_matrix, labels)

    # Information Gain (using f_classif)
    ig_scores, _ = f_classif(feature_matrix, labels)

    # Document Frequency (DF)
    df_scores = np.array((feature_matrix > 0).sum(0)).squeeze()

    # TF-IDF
    tfidf_scores = np.array(tfidf_vectorizer.idf_)

    # Creating a dictionary to store feature scores
    feature_scores = {
        #"Mutual Information": mi_scores,
        "Chi-Square": chi2_scores,
        "Information Gain": ig_scores,
        "Document Frequency": df_scores,
        "TF-IDF": tfidf_scores
    }

    # Ranking features based on each method
    ranked_features = {}
    for method, scores in feature_scores.items():
        ranked_indices = np.argsort(scores)[::-1]
        ranked_features[method] = [(count_vectorizer.get_feature_names_out()[i], scores[i]) for i in ranked_indices]

    return ranked_features

# Ranked features
ranked_features = rank_features(count_features, numerical_labels)

# Printing the ranked features for each method
for method, features in ranked_features.items():
    print(f"Ranking based on {method}:")
    for feature, score in features:
        print(f"{feature}: {score:.4f}")
    print("\n")
```

Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
# You code here (Please add comments in the code):

# importing and installing the necessary packages and libraries

from transformers import BertTokenizer, BertModel
import torch
from sklearn.metrics.pairwise import cosine_similarity

# Loading pre-trained BERT model and tokenizer

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Sample text data from Question 2 along with few more texts
texts = [
    "The product is really good. I love it! Not bad, but could be better. ğŸ˜Š",
    "This product is terrible. I hate it!",
    "It's an okay product, not great.",
]

# Design your query
query = "This product is excellent."

# Tokenizing and encoding the query
query_tokens = tokenizer(query, padding=True, truncation=True, return_tensors="pt")
query_embedding = model(**query_tokens).last_hidden_state.mean(dim=1)

# Encoding the text data and calculate cosine similarity
similarities = []
for text in texts:
    text_tokens = tokenizer(text, padding=True, truncation=True, return_tensors="pt")
    text_embedding = model(**text_tokens).last_hidden_state.mean(dim=1)
    similarity = cosine_similarity(query_embedding.detach().numpy(), text_embedding.detach().numpy())
    similarities.append(similarity[0][0])

# Ranking the text data based on similarity in descending order
ranked_indices = sorted(range(len(similarities)), key=lambda i: similarities[i], reverse=True)

# Printing ranked text
for idx in ranked_indices:
    print(texts[idx], "-", similarities[idx])
```
